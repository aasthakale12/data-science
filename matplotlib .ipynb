```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,6,11)
y=x**2
```

```python
# Create Figure (empty canvas)
fig = plt.figure()
```

```python
# Add set of axes to figure
axes = fig.add_axes([0.01, 0.01, 0.3, 0.3]) #left , bottom, width, height (range 0 to 1)
```

```python
#plot on that set of axes. 
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # notice that use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
x=np.linspace(0,6,11)
y=x**2
#Create Figure (empty canvas)
fig=plt.figure()
#Add set of axes to figure
axes=fig.add_axes([0.01,0.01,0.5,0.5])#left,bottom,width,height(range 0to1)
axes1=fig.add_axes([0.1,0.25,0.2,0.2])

#plot on that set of axes
axes.plot(x,y,'b--')
axes.set_xlabel('set X Label')#Notice the use of set_ to begin methods
axes.set_ylabel('set Y Label')
axes.set_title('set Title')

axes1.plot(y,x,'r*-')
axes1.set_xlabel('set X Label')#Notice the use of set_ to begin methods
axes1.set_ylabel('set Y Label')
axes1.set_title('set Title')
```

```python
fig1,ax=plt.subplots(2,3,sharex=True,sharey=True,layout='constrained')

ax[1,2].plot(x,y)

ax[0,1].plot(y,x)
```

Jointplot

jointplot() allows youn to basically match up two distplots for bivariate data. with your choice of what kind parameter to compare with:

    > "scatter"
    > "reg"
    > "resid"
    > "kde"
    > "hex"

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```
